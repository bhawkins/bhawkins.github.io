# Bad Data Masking

Some AirMOSS flights were plagued by data dropouts.  We could fairly easily process the data in segments, but the data backend system does not have a way to track multiple segments of the same line.  Another solution that would avoid the bookeeping problem is to blindly process the entire data file and then apply a mask to the output image.  To generate the mask, we need to know how to map regions in the raw data domain to regions in the image domain.  The relevant relationship is the imaging time equation, 

$${\hat n} \cdot {\hat l} = \sin (\alpha)$$

where $\alpha$ is the azimuth angle, ${\hat l}$ is the unit look vector and ${\hat n}$ is the image plane normal defined as

$$\hat{n} = \begin{bmatrix} \cos \theta_p \cos \theta_y \\\\ -\cos \theta_p \sin \theta_y \\\\ \sin \theta_p \end{bmatrix}$$

## Simulation Parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos

c = 2.99792458e8
window_parameter = 0.886

resolution = 1.0
wavelength = c / 1.2757e9
yaw = np.radians (10.0)
pitch = np.radians (2.0)
esa = np.radians (-8.0)
h = 12500.

nr = 9900
dr = 1.6655136
r0 = 13e3
r = r0 + dr * np.arange (nr)
```

## Angular Extent

```python
cosbeta = (h/r + sin(esa) * sin(pitch)) / (cos(esa) * cos(pitch))
sinbeta = sqrt (1.0 - cosbeta**2)  # positive for left-looking radar
sq = np.arcsin (sin(esa)*cos(pitch)*cos(yaw) 
                + cos(esa)*sin(yaw)*sinbeta 
                + cos(esa)*sin(pitch)*cos(yaw)*cosbeta)
beamwidth = 2 * np.arcsin (window_parameter * wavelength 
                           / (4 * resolution * np.cos (sq)))
sqmax = sq + beamwidth / 2
sqmin = sq - beamwidth / 2
```

```python
rkm = r/1000
plt.fill_between (rkm, np.degrees (sqmin), np.degrees (sqmax), facecolor='gray')
plt.plot (rkm, np.degrees (sq))
plt.xlabel ('Range [km]')
plt.ylabel ('Squint [degrees]');
```

## Spatial Extent

```python
s0min = r * sin(sqmin)
s0max = r * sin(sqmax)
plt.fill_between (rkm, s0min/1000, s0max/1000, facecolor='gray')
plt.xlabel ('Range [km]')
plt.ylabel ('S0 [km]');
```

```python
smin = s0min - r * sin(sq)
smax = s0max - r * sin(sq)
plt.fill_between (rkm, smin/1000, smax/1000, facecolor='gray')
plt.xlabel ('Range [km]')
plt.ylabel ('S [km]');
```

## Simple Model

```python
ssmin = -r * sin (beamwidth/2)
ssmax = r * sin (beamwidth/2)

plt.fill_between (rkm, ssmin/1000, ssmax/1000, facecolor='gray')
plt.xlabel ('Range [km]')
plt.ylabel ('S [km]')
plt.figure()
plt.plot (rkm, ssmin - smin)
plt.xlabel ('Range [km]')
plt.ylabel ('S Error [m]');
```
